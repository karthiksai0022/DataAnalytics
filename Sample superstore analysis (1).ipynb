# HANDLING OUTLIERS :

```python
num_data=df.select_dtypes(exclude=['object',bool])
```

```python
cat_data=df.select_dtypes(include=['object',bool])
```

```python
for i in num_data.columns:
    sns.boxplot(x=df[i])
    plt.show()
```

```python
col1=list(num_data.columns)
```

```python
def remove_outliers(cols):
    for i in cols:
        q1=df[i].describe()['25%']
        q3=df[i].describe()['75%']
        iqr=q3-q1
        upperlimit=q3+1.5*iqr
        lowerlimit=q1-1.5*iqr
        df[i]=np.where(df[i]>upperlimit,upperlimit,df[i])
        df[i]=np.where(df[i]<lowerlimit,lowerlimit,df[i])
remove_outliers(col1)        
        
```

```python
for k in col1:
    sns.boxplot(x=df[k])
    plt.show()
```

# BARCHART: USING LIST

```python
import matplotlib.pyplot as plt
import seaborn as sns

categories = ['Ship Mode', 'Segment', 'Country', 'City', 'State', 'Region','Category', 'Sub-Category']
fig, axs = plt.subplots(nrows=4, ncols=2, figsize=(15, 25))
ax = axs.flatten()

for i, var in enumerate(categories):
    sns.countplot(x=var, hue='Region', data=df, ax=ax[i])
    ax[i].set_xticklabels(ax[i].get_xticklabels(), rotation=90)

fig.tight_layout()
plt.show()
```

# PIE CHART: USING LIST

```python
import matplotlib.pyplot as plt
import seaborn as sns

categories = ['Ship Mode', 'Segment', 'Country', 'City', 'State', 'Region','Category', 'Sub-Category']
fig, axs = plt.subplots(nrows=4, ncols=2, figsize=(20, 25))
ax = axs.flatten()

for i, var in enumerate(categories):
    if i<len(axs.flat):
        cat_counts=df[var].value_counts()
        axs.flat[i].pie(cat_counts,labels=cat_counts.index,autopct='%.2f',startangle=90)
        axs.flat[i].set_title(f'{var} Distribution')
    
fig.tight_layout()
plt.show()
```
